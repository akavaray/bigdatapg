평균 = (a + b + c) / 3  
=(x_1 + x_2 + x_3) / 3

```python
x = [10,20,30]
sum = 0
for i in range(3):
    sum += x[i]
print("평균:", sum / len(x))
```

분위수: 데이터를 순서대로 늘어 놓았을 때 균등한 간격으로 나누는 기준점  
이분위수: 중앙값 (데이터를 두 구간으로 나누기에)  
사분위수: 25%, 50%, 75% (네 구간으로 나눔)

quantile(): 분위수 계산

```python
ns_book7["대출건수"].quantile(0.25)
```

```python
ns_book7["대출건수"].quantile([0.25,0.5,0.75])
```

quantile() 각 분위수에 해당하는 값을 담은 시리즈 객체 반환  
시리즈 객체 정의 후 분위수 계산

```python
pd.Series([1,2,3,4,5]).quantile(0.9)
```

interpolation 매개변수에서 중간 값을 계산  
매개변수를 따로 지정하지 않으면 기본 값으로 계산  
interpolation(보간): 두 지점 사이의 특정 위치의 값을 구하는 방법

```python
4 + (0.9 - 0.75) * (5 - 4) / (1.0 - 0.75)
```

위의 방식은
interpolation 매개변수 "linear",   
"midpoint": 분위수 상관 없이 두 수 사이의 중앙값 사용

```python
pd.Series([1,2,3,4,5]).quantile(0.9, interpolation="midpoint")
```

"nearest": 두 수 중에서 가까운 값

```python
pd.Series([1,2,3,4,5]).quantile(0.9,interpolation="nearest")
```

데이터프레임에서 기술통계 구하기  
numeric_only = True 로 지정해야함  
지정하지 않으면 모든 데이터 타입의 열에 대해 수행하기에 시간이 매우 오래 걸리고 경고 발생

```python
ns_book7.mean(numeric_only=True)
```

ns_book7 번호 열은 모두 고유한 값이기에 최빈값이 의미가 없음  
loc 메서드로 '도서명' 열부터 마지막 열까지 최빈값 찾기

```python
ns_book7.loc[:,'도서명':].mode()
```

도서명 열에 가장 많이 등장하는 값은 승정원 일기이고 출판사 열에 가장 많이 등장하는 값은 문학동네다.  
결과값만 보고 승정원일기가 문학동네 출판사의 도서라고 오해하면 안됨  
**mode() 출력 결과물 사이에는 서로 연관 X**
